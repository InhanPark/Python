```python
import statistics as st  # statistics module 은 통계분석을 위한 다양한 함수를 제공한다.

# 숫자의 모임인 data 를 얻으면 그것의 중심특성을 먼저 파악하여야 한다.
# 중심 특성에는 mean(평균), median(중앙값), mode(최빈값) 이 있다.
# 세가지 값을 구해서 이 숫자모임들이 어떤 경향을 갖는지 파악한다.

# 그 다음에는 숫자들이 이 중심값에 얼마나 가까이 있는지를 파악해야 한다.
# 숫자들이 중심값에서 많이 떨어져 있다면 그 값으로 숫자들을 표현할 수 없다.
# 예를들어 어느 반이 95명이 0점이고, 5명이 100점이면 중심값을 가지고
# "이 학급이 공부를 못한다." 라고 결론 내기는 어렵다.

# 이런 문제 때문에, 숫자들이 중심값에서 얼마나 떨어져 있다는 반드시 봐야 한다. 
# 이를 dispersion 이라고 한다. Dispersion 에는 몇가지 표현 방법이 있다.  

# data 를 list 에 넣어서 분석하면 매우 편하다
scores = [13, 35, 47, 29, 61, 52, 34, 46, 58, 20, 46]

# 먼저 data 의 갯수가 통계 분석을 할 정도로 충분한지 본다.
print('count: ', len(scores)) 


# 먼저 중심특성을 본다.
print('mean: ', st.mean(scores))   # 평균값
print('median: ', st.median(scores)) # 중앙값
print('mode: ', st.mode(scores))   # 최빈값

# Population Variance (모집단 분산)
# 각 값에서 평균치를 뺀다. 그다음 각 값을 제곱하여 더한다.
# 제곱을 하는 이유는 중심치에서 떨어진 정도를 (+/- 관계없이) 보아야 하기 때문이다.
# 아래의 함수로 쉽게 구할 수 있다.
print('p variance :', st.pvariance(scores))

# Population Standard Deviation (모집단 표준편차)
# Population Variance 를 제곱근 계산 한 것이다. 이렇게 하면 특이 data (outlier) 가 전체를 왜곡하는 일을 줄일 수 있다.
# 또한 제곱근을 하면 원래 data 와 단위가 같아진다 (제곱을 하면 단위도 제곱)
print('p stddev :', st.pstdev(scores))
```
