```python
"""Boolean Algebra, Ture 와 False 의 결정"""

# Program code 는 control 의 flow 이다. 이 flow 는 중간중간에서 항상 흐름의 방향을 결정하는 부분이 나타난다.
# 이 결정은 True, False 의 두가지중 하나를 결정하는 것이고, 그것을 다루는 것이 Boolean Algebra 이다.

# 비교(comparison)를 하면 True, False 를 얻게됨
print(7 > 4)
print(7 > 4, 7 < 4)
# 비교하는 방법은 6가지가 있음: ==, !=, <, >, <=, >=
# 연산자 우선 순위는 () -> ** -> * / // % -> + - -> > <= < >= -> == !=
```

```python
# 비교 결과를 가지고 if 문을 구성 하면됨
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))

if number1 == number2:
  print(number1, 'is equal to', number2)

if number1 != number2:
  print(number1, 'is not equal to', number2)

if number1 < number2:
  print(number1, 'is less than', number2)

if number1 > number2:
  print(number1, 'is greater than', number2)

if number1 <= number2:
  print(number1, 'is less than or equal to', number2)

if number1 >= number2:
  print(number1, 'is greater than or equal to', number2)
```

```python
# Boolean operator 'and', 'or', 'not' 을 사용하면 좀 더 복잡한 조건을 만들 수 있음

# and: 두 조건이 모두 True 일때만 True
gender = 'Female'
age = 70

if gender == 'Female' and age >= 61:
  print('Senior female')

# or: 두 조건이 모두 False 일때만 False
semester_average = 83
final_exam = 95

if semester_average >= 90 or final_exam >= 90:
  print('Student gets an A')

# not: True 이면 False 로 False 면 True 로
grade = 87

if not grade == -1:
  print('The next grade is', grade)
```
