```python
"""Repetition Flow (반복 흐름) 의 Control, FOR & WHILE 문"""

# if 문은 흘러가는 (Sequential) 것을 control 하는데 반해서,
# 반복하도록 (Repetition) control 하는것이 for, while 문이다.
# python 은 do~while 은 지원하지 않는다
```

```python
# while 문은 조건이 True 인 동안은 계속 반복되는 것이다.
product = 3

while product <= 50:
  product = product * 3
  print (product)
```

```python
# 이렇게 쓰면 영원히 돌아간다. ('Runtime > Interrupt execution' 으로 중지)
while True:
  product = product * 3
  print (product)

# 이렇게 쓰면 영원히 안돌아간다.
while Flase:
  product = product * 3
  print (product)
```

```python
# python 의 for 문은 다른 언어와 좀 다르다
# 'in' 뒤에 있는것을 한개씩 가져와서 실행하는 방식이다
# 익숙해지면 훨씬 편하다

# 여기에는 아래예와 같이 string 을 놓을수도 있지만, 다른것도 가능하다
for character in 'Programming':
  print(character)
```

```python
# list 라는 data type 를 사용해서 이렇게 할수도 있다
number = 1

for name in [2, -3, 'Elin', 0, 17, 'Inhan']:
  print(number, name)
  number = number + 1
```

```python
# range built-in 함수를 사용하면 C 와 비슷하게 동작한다.
# range 는 start, stop, step 을 지정할 수 있다.
sum = 0

for number in range(1000):
  sum = sum + number

print(sum)
```
