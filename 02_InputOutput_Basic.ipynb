```python
"""Input & Output with console"""

# input 은 data 를 저장할 변수에 prompt 와 같이 쓴다
name = input("What's your name? ")
print("stored name is: ", name)
```

```python
# input 받은 data 는 언제나 string 이다.
value1 = input("Enter first number: ")
value2 = input("Enter second number: ")
value3 = value1 + value2
print(value1, value2, value3)
print(type(value1), type(value2), type(value3))

# 숫자로 쓰려면 type 변환을 해야한다.
value4 = int(value1) + int(value2)
print(value1, value2, value3, value4)
print(type(value1), type(value2), type(value3), type(value4))
```

```python
# 받을때부터 숫자로 받는법.
value5 = int(input("Enter first number: "))
value6 = int(input("Enter second number: "))
value7 = value5 + value6
print(value5, value6, value7)
print(type(value5), type(value6), type(value7))
```

```python
# single- 과 double- quoted string 모두 print 가능
print('Welcome to Python!')
print("Welcome to Python!")
```

```python
# comma 로 분리된 string
print('Welcome', 'to', 'Python!')
```

```python
# Escape sequence 로 print
print('Welcome\nto\n\nPython!') # '\n' 은 next line
print('Welcome\tto\t\tPython!') # '\t' 은 tab
print('Welcome\\to\\Python!') # '\\' 은 '\' 표시
print('Welcome\'to\' \"Python!\"') # '\'' 은 ''' , '\"' 은 '"'
```

```python
# 긴라인을 표시하는 방법. 한줄로 나타남
print('this is a longer string, so we \
split it over two lines')
```

```python
# String 과 숫자를 표시하는 방법. 그냥 나열하면 됨
a = 3
print('Sum is', 7 + a, '변수 a 는 ', a)
```
