```python
"""Keyword 로 사용하는 Data, Dictionary"""

#사전(dictionary)에서 'code' 라는 단어를 찾아보면, 아래와 같이 나온다.
# code: 1. 상사(商社)가 국제 전보에서 정하여 두고 쓰는 약호·기호·암호. 순화어는 `기호', `약호'. 2. 정보를 나타내기 위한 기호의 체계
#'code' 라는 단어를 몇번째 단어라고 순서로 말하려하면 무척 어려울 것이다. 그보다는 'code'라는 key 에 '설명' 이라는 내용이 있다고 생각하면 쉬울 것이다.
#'simple_dic' 이라는 나의 사전을 만들어보자. 그냥 name = {'key':'item', 'key':'item', ...} 이렇게 입력하면 된다.
#찾을때는 'name.get('key')' 로 하면 된다.

simple_dic = {'code': '1. 상사(商社)가 국제 전보에서 정하여 두고 쓰는 약호·기호·암호. 순화어는 ''기호'', ''약호''.\n2. 정보를 나타내기 위한 기호의 체계', 'apple': '원영이의 별명', 'audi': '아빠의 10년 넘은 SUV'}

print(simple_dic.get('apple'))
```

```python
"""이론적 설명 (중요), Data Structure"""

#Data Structure 는 data 를 저장하는 방법이다. 대표적인 것은 3가지이다.
# 1. List Family: list, array, stack, queue, string, tuple
# 2. Tree Family: tree, heap, hash table
# 3. Graph Family: graph 

#C/C++ 은 이런 structure 를 만들 수 있는 방법을 제공한다. 만들어 쓰는 것은 user 가 한다. Python 은 아예 이런 structure 와 사용방법을 제공한다 (C 로 만들어져 있음. Python 으로 만들면 엄청 느려짐). User 는 잘 쓰기만 하면된다.

#List Family 는 순서(sequence)가 있다. 이것은 순서를 기준하여 사용할 수 있다는 것이다. 별도로 정해주지 않아도 각 item 이 순서라는 번호 data 를 value 와 같이 가지고 있다는 것. 이들은 '다음 data', '이전 data', '앞에서 몇번째 data', '지금것 기준 몇번째 data' 라는 식으로 사용할 수 있다.

#Tree Family 는 순서를 기준으로 data 를 저장하지 않고, 어떤 관련성을 기준으로 한다 (예를들면 key 라는 이름). 사전에서 단어를 찾을때 첫단어부터 순서대로 찾는 사람은 없다. 사전을 펼쳐보면 내가 찾는 단어가 그 앞에 있는지 뒤에 있는지 알수있다. 그런 방법으로 접근하여 단어를 찾는게 빠르다. 이것이 tree 구조의 원리이다.

#Graph 는 item 들이 서로 얽혀있어서 이런식으로 찾기 어려운 경우이다. 친구사이에 서로 친한 정도를 표시하려면 list 나 tree 는 어렵다. 또한 인간을 포함한 동물의 기억 data 는 list 나 tree 가 아닌 graph 로 저장되어 있다. 이런 data 를 다루는 대표적인 것이 deep learning, tensorflow 이다. 
```

```python
"""Dictionary Data Structure 다루기"""

#로마숫자를 아라비아숫자로 바꾸는 program 을 만들기 위해 아래와 같은 dictionary data 를 만들었다
roman_numerals = {'I': 1, 'II': 2, 'III': 3, 'V': 5, 'X': 100}

#Data length 구하고, 전체 data 보기
print(len(roman_numerals), roman_numerals)

#Data 꺼내쓰기
for roman, arabic in roman_numerals.items():
    print(f'{roman} corresponds to {arabic}')

print(roman_numerals.get('V'))
print(roman_numerals['X'])

#Value 의 내용 바꾸기 (Key 내용은 못바꿈)
roman_numerals['X'] = 10
print(len(roman_numerals), roman_numerals)

#Data (key:value) 추가하기
roman_numerals['L'] = 50
print(len(roman_numerals), roman_numerals)

#Data (key:value) 지우기
del roman_numerals['III']
print(len(roman_numerals), roman_numerals)

#Data value 를 얻고, 지우기
print(roman_numerals.pop('X'))
print(len(roman_numerals), roman_numerals)

#없는 data 를 찾기 (get 을 쓰면 error 가 나지 않고, None 를 return)
#print(roman_numerals['III'])
print(roman_numerals.get('III'))

#Data 가 있는지 확인하기
print('V' in roman_numerals)
print('III' in roman_numerals)
```
