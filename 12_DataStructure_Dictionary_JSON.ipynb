```python
"""Keyword 로 사용하는 Data, Dictionary"""

#사전(dictionary)에서 'code' 라는 단어를 찾아보면, 아래와 같이 나온다.
# code: 1. 상사(商社)가 국제 전보에서 정하여 두고 쓰는 약호·기호·암호. 순화어는 `기호', `약호'. 2. 정보를 나타내기 위한 기호의 체계
#'code' 라는 단어를 몇번째 단어라고 순서로 말하려하면 무척 어려울 것이다. 그보다는 'code'라는 key 에 '설명' 이라는 내용이 있다고 생각하면 쉬울 것이다.
#'simple_dic' 이라는 나의 사전을 만들어보자. 그냥 name = {'key':'item', 'key':'item', ...} 이렇게 입력하면 된다.
#찾을때는 'name.get('key')' 로 하면 된다.

simple_dic = {'code': '1. 상사(商社)가 국제 전보에서 정하여 두고 쓰는 약호·기호·암호. 순화어는 ''기호'', ''약호''.\n2. 정보를 나타내기 위한 기호의 체계', 'apple': '원영이의 별명', 'audi': '아빠의 10년 넘은 SUV'}

print(simple_dic.get('apple'))
```

```python
"""Dictionary Data Structure 기본 다루기"""

#로마숫자를 아라비아숫자로 바꾸는 program 을 만들기 위해 아래와 같은 dictionary data 를 만들었다
roman_numerals = {'I': 1, 'II': 2, 'III': 3, 'V': 5, 'X': 100}

#Data length 구하고, 전체 data 보기
print(len(roman_numerals), roman_numerals)

#Data 꺼내쓰기
for roman, arabic in roman_numerals.items():
    print(f'{roman} corresponds to {arabic}')

print(roman_numerals.get('V'))
print(roman_numerals['X'])

#Value 의 내용 바꾸기 (Key 내용은 못바꿈)
roman_numerals['X'] = 10
print(len(roman_numerals), roman_numerals)

#Data (key:value) 추가하기
roman_numerals['L'] = 50
print(len(roman_numerals), roman_numerals)

#Data (key:value) 지우기
del roman_numerals['III']
print(len(roman_numerals), roman_numerals)

#Data value 를 얻고, 지우기
print(roman_numerals.pop('X'))
print(len(roman_numerals), roman_numerals)

#없는 data 를 찾기 (get 을 쓰면 error 가 나지 않고, None 를 return)
#print(roman_numerals['III'])
print(roman_numerals.get('III'))

#Data 가 있는지 확인하기
print('V' in roman_numerals)
print('III' in roman_numerals)
```

```python
"""Key 와 Value 를 이용한 Dictionary Method"""

months = {'January': 1, 'February': 2, 'March': 3}

# keys 와 values method 는 list 와 유사한 data view 를 제공한다.
print(months.keys())
print(months.values())
print(months.items())
for month_name in months.keys():
    print(month_name, end=' ')

# 이것은 list 로 변환할 수도 있다.
print(list(months.keys()))

# sort를 할 수도 있다.
print(sorted(months.keys()))

# dictionary 두개가 같은지도 비교할 수 있다.
months3 = {'March': 3, 'February': 2, 'January': 1}
months4 = {'January': 1, 'February': 2, 'Mar': 3}

print(months == months3)
print(months == months4)

# 내용을 update 할 수 있다
months.update({'April': 5}) # item 을 추가한다
print(months.items())
months.update(April = 4)    # value 를 update 한다
print(months.items())

# dictionary 에서 또다른 dicstionary 를 만들 수 있다. (Dictionary Comprehensions)
grades = {'Sue': [98, 87, 94], 'Bob': [84, 95, 91]}
grades2 = {k: sum(v) / len(v) for k, v in grades.items()}
print(grades.items())
print(grades2.items())

# list 나 range 에서도 만들 수 있다.
num_pair = {number: number ** 3 for number in range(1, 6)}
print(num_pair.items())
```

```python
"""JSON (Java Script Object Notation)"""
# JSON 은 cloud application 사이에 communication 을 할때 많이 사용하는 format 이다. 구조는 dictionary 와 같다. (key:value 대신 property names:values 로 표현만 다르다)
# value 에 사용할 수 있는 data 는 string (" "), numbers (100, 24.98), boolean (true, false), null (python 은 None), array ([100, 200, 300]), 그리고 다른 JSON object 들이다.

# dictionary data 하나를 가정하자.
accounts_dict = {'accounts': [{'account': 100, 'name': 'Jones', 'balance': 24.98}, {'account': 200, 'name': 'Doe', 'balance': 347.67}]}

# 이것을 JSON format file 로 바꾸려면 json.dump method 를 사용한다.
import json

with open('accounts.json', 'w') as accounts:
    json.dump(accounts_dict, accounts)
    
# 이러한 file 이 만들어졌다. String 이 double-quote 가 되었다.
# {"accounts": [{"account": 100, "name": "Jones", "balance": 24.98}, {"account": 200, "name": "Doe", "balance": 347.67}]}
# 이 과정을 data 의 serializing 이라고 한다.
```

```python
# 반대로 JSON format 을 읽어들여서 Python format 으로 바꾸는 것은 data 의 deserializing 이라고 한다. jason.load method 를 사용한다.

with open('accounts.json', 'r') as accounts:
    accounts_json = json.load(accounts)

# 이제 dictionary data 로 사용할 수 있다.    
print(accounts_json)
print(accounts_json['accounts'])
print(accounts_json['accounts'][0])
print(accounts_json['accounts'][1])
```

```python
# JSON data 를 읽어들여서 좀 더 가독성 있게 보려면 아래와 같이 한다.
with open('accounts.json', 'r') as accounts:
    print(json.dumps(json.load(accounts), indent=4))
```
