```python
"""List 기본 사용법"""

#List 는 'name = [ items ]' 로 간단히 만들면 됨
c = [1, 3, 5, 7, 9]

#range 를 사용해서 만들수도 있음
d = list(range(11, 29, 2)) #range(start, stop, step) stop 숫자는 안만들어짐
#range 사용에는 많은 방법이 있으니 필요시 인터넷 검색하여 보면 됨

#'len(name)'으로 length 구할 수 있음
print(len(c), c)

#각 item 들은 c[0] ~ c[길이-1] 로 index 할 수 있다.
print(c[0], c[4])

#negative number 로도 index 가능하다. c[-1] 은 c[길이-1] 이다
print(c[-1], c[-2], c[-3], c[-4], c[-5])

#index 를 연산할 수 있다. (단 integer 로)
print(c[3-2])

#list 를 합쳐서 새로운 것을 만들수도 있다 (concatenation)
c += d
print(len(c), c) #length 자체가 변했음

#for 문을 사용해서 모든 index 에 접근 할 수 있다.
for i in range(len(d)):
    print(f'{i}: {d[i]}')
    
```

```python
"""필요에 따라 만들고, 없애고, 늘이고, 줄이고"""

#element 추가하는 방법 (list_name += [element])
a_list = []        #empty list
a_list += [6]      #int 를 추가
a_list += ['Hi']   #string 을 추가
a_list += [87.4]   #float 을 추가
a_list += 'Python' #이렇게 하면 각 char 을 추가하게 됨
print(len(a_list), a_list)
e = a_list[::2]    #기존 list 에서 slicing (오려내기) 하여 만들수도 있음
print(len(e), e)
#list slicing 은 많은 방법이 있으니 필요시 인터넷 검색하여 보면 됨

#element 뺴는 방법
del a_list[2]      #2번 element (87.4) 를 삭제 
print(len(a_list), a_list)  #length 자체가 변했음
del a_list[2:3]    #2~3번 element 를 범위지정 삭제
print(len(a_list), a_list)  #length 자체가 변했음
del a_list[:]      #모든 element 를 삭제
print(len(a_list), a_list)  #length 자체가 변했음
del a_list         #더이상 안쓰면 지울수도
print(len(a_list), a_list)  #지워졌기에 에러남!!!
```

```python
"""Program 끝날때까지 바뀌면 안되는 List, Tuple"""

#Program 실행의 기초 data 로 내용이 바뀌면 안되는 것들은 Tuple 에 저장한다.
#List 는 'name = [ items ]' 인데, Tuple 은 'name = ( items )' 로 괄호만 틀리다.

student_tuple = ()  #Tuple 은 괄호로 감싼다

student_tuple = 'John', 'Green', 3.3      #comma 로 분리하여 입력한다.
print(len(student_tuple), student_tuple)

student_tuple = ('Lexy', 'Hamilton', 4.3) #괄호로 둘러싸도 된다.
print(len(student_tuple), student_tuple)

student_tuple += ('John', 'Green', 3.3)   #추가는 list 와 같이 한다
print(len(student_tuple), student_tuple)

student_tuple += ('Elin',)                #한개만 추가할때는 반드시 comma 를 붙임
print(len(student_tuple), student_tuple)

student_tuple += ('Park', 4.7)            
print(len(student_tuple), student_tuple)

student_tuple += ('Micky', 'Mouse', [98, 85, 87]) #List 도 추가할 수 있다.
print(len(student_tuple), student_tuple)

student_tuple[11][2] = 77                #이것은 list 이므로 data 를 바꿀 수 있다
print(len(student_tuple), student_tuple)

student_tuple[0] = ('Tom',)               #Tuple data 는 바꾸지 못한다.
```

```python
"""List/Tuple 에 저장해 놓은것 꺼내쓰기 (Unpacking Sequences)"""
#data 를 아래와 같이 list 나 tuple 로 묶을 수 있다 (packing)
number1 = 99
number2 = 22
number3 = 77
number_L = [number1, number2, number3]
number_T = (number3, number2, number1)
print(number_L, number_T)

#풀어서 꺼낼때는 반대로 하면된다.
num_1, num_2, num_3 = number_L;
print(num_1, num_2, num_3)
num_1, num_2, num_3 = number_T;
print(num_1, num_2, num_3)

#왼쪽의 변수가 한개이면 풀지를 않고 그대로 넘겨준다
#꺼낼때 왼쪽의 변수 갯수와 오른쪽의 item 갯수가 다르면 error 가 발생한다.
num_1 = number_L; #이건 그대로 list 로 나옴
print(num_1)
#num_1, num_2 = number_L; #이건 error 로 나와서 comment 처리함
#print(num_1, num_2)

#실제로는 Group 으로 data 를 packing 해서 저장하고, unpacking 해서 사용함
group_data = [[1, 3], [4, 5], [89, 74]] #list 를 이용하여 group 으로 data 저장
for num_1, num_2 in group_data:  #for 문을 이용하여 차례로 꺼내어 사용
    print(num_1, num_2)
```

```python
"""List/Tuple 의 각 item 에 번호를 붙여주기 (enumeration)"""

#숫자의 list 를 만들고
numbers = [19, 3, 15, 7, 11]
print(len(numbers), numbers)

#enumerate()를 적용하면 각 value 에 index 가 생김
for index, value in enumerate(numbers):
    print(index, value)

#이것은 list 가 아니고 별개 object 임    
print(enumerate(numbers))

#이렇게 응용하면 bar chart 를 그릴수도 있음
for index, value in enumerate(numbers):
    print(f'{index:>5}{value:>8} {"@" * value}')
```

```python
"""List/Tuple 에서 필요한 부분을 잘라내어 새로운 List/Tuple 만들기"""

numbers = [2, 3, 5, 7, 11, 13, 17, 19]

#범위를 정해서 잘라내기
new_num = numbers[2:6]
print(len(new_num), new_num)

#끝만 정해서 잘라내기
new_num = numbers[:6]
print(len(new_num), new_num)

#index 가 없는 경우, 전체를 다 잘라냄
new_num = numbers[:]
print(len(new_num), new_num)

#index step 을 정해서 잘라내기
new_num = numbers[::2]
print(len(new_num), new_num)

#negative index step 으로 잘라내기 -> 거꾸로 나오게 됨
new_num = numbers[::-1]
print(len(new_num), new_num)
new_num = numbers[-1:-9:-1]
print(len(new_num), new_num)

#index를 주어서 일부 내용 변경하기
numbers[0:3] = ['two', 'three', 'five']
print(len(numbers), numbers)

#index를 주어서 일부 내용 삭제하기
numbers[0:3] = []
print(len(numbers), numbers)

#index를 주어서 하나건너 내용 변경하기
numbers = [2, 3, 5, 7, 11, 13, 17, 19]
numbers[::2] = [100, 100, 100, 100]
print(len(numbers), numbers)

#del 명령으로 일부 내용 삭제하기
print("... del command ...")
numbers = list(range(0, 10))
print(len(numbers), numbers)
del numbers[-1]
print(len(numbers), numbers)
del numbers[0:2]
print(len(numbers), numbers)
del numbers #numbers 객체를 아예 삭제
```

```python
"""Function 에 List 를 parameter 로 전달하기 (List 처리 함수를 마음대로 만듦)"""

def modify_elements(items): #list 는 reference 로 전달되므로 내용이 바뀐다
    """Multiplies all element values in items by 2."""
    for i in range(len(items)):
        items[i] *= 2
        
numbers = [10, 3, 7, 1, 9]
modify_elements(numbers)
print(len(numbers), numbers)
```

```python
"""List 를 sorting 해서 원본을 변경하거나, 새로운 list 만들기"""

numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

sorted_num = sorted(numbers) #이렇게 하면 sorted_num 이 새로 만들어진다
print(len(numbers), numbers)
print(len(sorted_num), sorted_num)

numbers.sort() #이렇게 하면 원본 'numbers' 자체가 sort 된다
print(len(numbers), numbers)

numbers.sort(reverse=True) #이렇게 하면 원본 'numbers' 자체가 reverse sort 된다
print(len(numbers), numbers)
```

```python
"""List/Tuple 에서 찾기 (Searching)"""

#단어들을 저장해둔 list
vocabulary = ['apple', 'sunflower', 'melon', 'rose', 'dog', 'cat']
vocabulary_2 = vocabulary * 2  #vocabulary 의 내용이 두번 들어있는 list
print(vocabulary)
print(vocabulary_2)

#vocabulary 에서 원하는 단어가 있는 index 찾기
print(vocabulary.index('dog'))

#vocabulary_2 의 index 6 에서 시작해서 원하는 단어가 있는 index 찾기
print(vocabulary_2.index('dog', 6))

#vocabulary 에서 'dog' 를 찾으면 첫번째것의 위치를 알려준다
print(vocabulary_2.index('dog'))

#vocabulary_2 의 index 5 에서 끝까지 사이에서 'dog' 단어를 찾기
print(vocabulary_2.index('dog', 5, len(vocabulary_2)))

#vocabulary 안에 'sunflower' 가 있는지, 없는지를 각각 확인함
print('sunflower' in vocabulary)
print('sunflower' not in vocabulary)

#if 문을 사용하면 이런 응용을 할 수 있음 (단어 찾기)
find_word = 'carrot'

if find_word in vocabulary:
    print(f'found {find_word} at index {vocabulary.index(find_word)}')
else:
    print(f'{find_word} not found')
```

```python
"""List 에 적용하는 methods"""

color_names = ['orange', 'yellow', 'green']
print(len(color_names), color_names)

#위치를 지정해서 new item 삽입하기
color_names.insert(0, 'red') #index 0 위치에 'red' 삽입
print(len(color_names), color_names)

#맨끝에 new item 붙이기
color_names.append('blue') #color_names += ['blue']와 효과 같음
print(len(color_names), color_names)

#맨끝에 다른 list 붙이기
color_names.extend(['indigo', 'violet']) #color_names += ['indigo', 'violet']와 효과 같음
print(len(color_names), color_names)

#item 삭제하기 (같은것이 2개 이상이면 첫번째 것)
color_names.remove('green') #color_names += ['blue']와 효과 같음
print(len(color_names), color_names)

#어떤 item 의 갯수 확인하기
color_names += ['blue']
print(len(color_names), color_names)
print(color_names.count('blue')) #color_names[:] = []와 효과 같음

#list 순서 바꾸기
color_names.reverse() #color_names[:] = []와 효과 같음
print(len(color_names), color_names)

#list 복사하기 (deep copy)
color_names_cp = color_names.copy() #color_names_cp = color_names[:]와 효과 같음
#C 의 address 값 구하기(&) 와 같은 기능이 id(변수명) 함수임
#copy 된것의 address 가 다르므로 deep copy (이름만 두개가 아니라 내용도 두개로) 임
print(id(color_names), color_names)
print(id(color_names_cp), color_names_cp)

#list 비우기
color_names.clear() #color_names[:] = []와 효과 같음
print(len(color_names), color_names)
print(len(color_names_cp), color_names_cp)
#color_names 를 clear 해도 color_names_cp 는 안바뀜. 서로 독립적임

#list 맨앞 혹은 맨끝의 item 을 뺴냄
color_names = ['violet', 'indigo', 'blue', 'yellow', 'orange', 'red']
print(len(color_names), color_names)
print(color_names.pop(0)) #맨앞의 item 을 뺴냄
print(len(color_names), color_names)
print(color_names.pop())  #맨끝의 item 을 뺴냄
print(len(color_names), color_names)
#이것을 맨뒤에 item 을 추가하는 append() method 와 같이 쓰면 queue (FIFO) 나 stack (FILO) 을 쉽게 만들 수 있다.
```
