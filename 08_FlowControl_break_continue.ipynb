```python
"""Class average program with sequence-controlled repetition."""
# 그냥 죽 처리하는 방법 (data 를 자동으로 받아서)

# initialization
total = 0  # sum of grades
grade_counter = 0
grades = [98, 76, 71, 87, 83, 90, 57, 79, 82, 94]
print(type(grades))
# 이것은 python 특유의 list 라는 data type 임. array 아님.

# prcessing phase
for grade in grades: # grades 에서 값을 하나씩 읽어서 grade 에 저장
  total += grade     # 합계에 추가함
  grade_counter += 1

# termination phase
average = total / grade_counter
print(f'Class average is {average}') # C 처럼 format 을 지정해 주고 print 하는 방법
```

```python
"""Class average program with sentinal-controlled iteration."""
# Sentinal(보초)를 두어서 입력되는 data 를 확인하며, 그에 따라 처리하는 방법

# initialization
total = 0  # sum of grades
grade_counter = 07

# prcessing phase
grade = int(input('Enter grade, -1 to end: ')) # get one grade

while grade != -1: # 이 부분이 sentinal
  total += grade     # 합계에 추가함
  grade_counter += 1
  grade = int(input('Enter grade, -1 to end: '))

# termination phase
if grade_counter != 0:
  average = total / grade_counter
  print(f'Class average is {average}') # C 처럼 format 을 지정해 주고 print 하는 방법
else:
  print('No grades were entered')
```

```python
"""Using nested control statements to analyze examination results."""
# 반복문(repetition) 안에 조건문(if)을 두어서 처리하는 방법

# initialize variables
passes = 0    # number of passes
failures = 0  # number of failures

# process 10 students
for student in range(10):
  # get one exam result
  result = int(input('Enter result (1=pass, 2=fail): '))

  if result == 1:
    passes += 1
  else:
    failures += 1

# termination phase
print('Passed: ', passes)
print('Failed: ', failures)

if passes > 8:
  print('Bonus to instructor')
```

```python
# break 는 어떤 조건이 맞으면 loop 을 빠져나옴

for number in range (3,100,2): # range 는 3에서 100까지 2씩 증가
  if number == 21:
    break
  print(number, end=' ')
```

```python
# continue 는 어떤 조건이 맞으면 loop 처음으로 가서 계속 실행

for number in range (3,30,2): # range 는 3에서 30까지 2씩 증가
  if number == 21:  # 값이 21 이면 이 다음을 실행하지 않고 처음으로 감
    continue
  print(number, end=' ')
```
