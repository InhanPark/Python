```python
"""Python Basic                               - Aug 01. 2024"""

"""기본 Data Type"""

#python 에서는 comment 를 하는 방법이 '#' 를 쓰는 방법 (comment) 와 '"""' 를 쓰는 방법 (Docstrings) 두가지 있음

#data type 지정은 매우 간단함. user 는 신경쓸 필요가 없음. 자동으로 처리
a = 3
b = 4.5
c = a + b

#data type 을 알고 싶으면 type 명령어를 사용하면 됨
print("실험 1:", a, b, c, type(a), type(b), type(c))

#data type 은 code 실행중에도 바뀜
a = a * 3.14
#dl 줄에서 변수 a 의 data type 이 int 에서 float 으로 바뀌었음
print("실험 2:",a, type(a))

#data type 을 function 으로 바꾸어 줄수도 있음
a = 2
b = 5.8
c = 5.8
print("실험 3:",a, b, c, type(a), type(b), type(c))

a = float(a)
b = int(b)   #int 는 소수점 이하를 버림
c = round(c) #float 는 소수점 이하를 반올림
print("실험 4:",a, b, c, type(a), type(b), type(c))
```

```python
"""Arithmetic Operations"""

#산술 연산은 7가지가 있음

print("Addition: ", 2 + 2)        #더하기
print("Subtraction: ", 100 - 75)  #빼기
print("Multiplication: ", 7 * 9)  #곱하기
print("Integer division: ", 15 // 2)  #정수 나눗셈 (소수점이하를 버림)
print("Floating-point division: ", 15 / 2)  #실수 나눗셈
print("Modulo: ", 15 % 2)         #나머지 구하기
print("Exponentiation: ", 2 ** 2) #제곱

#연산의 우선 순위는 일반 산수와 같음: () -> ** -> * // / -> + =
```

```python
"""돈계산용 특수 Data Type, Decimal"""

#소수점 아래에서도 정확히 떨어지는 숫자가 필요한 경우 float data type 으로는 안된다

from decimal import Decimal  #library 사용이 필요함

#float type 의 문제점. 소수점이하 20자리를 표시해보면 문제가 있다. 거액을 다룰때는 문제가 된다.
amount = 112.31
print(amount)
print(f'{amount:.20f}')

#decimal 을 사용하면 마치 정수를 다룰때처럼 깔끔하다
d_amount = Decimal('112.31')
print(f'{d_amount:.20f}')
```

```python
#Decimal 은 특수 data type 이고, 사칙연산이 가능하다

x = Decimal('10.5')
y = Decimal('2')

print( x + y)
print( x // y)

type(x)
```
