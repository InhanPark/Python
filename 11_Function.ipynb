```python
# 05_Data_Analysis_Basic 의 minimum 을 구하는 code 를 function 으로 만들기

def minimum(number1, number2, number3):
  """Return the maximum of three values.""" # function 설명
  minimum = number1
  if number2 < minimum:
    minimum = number2
  if number3 < minimum:
    minimum = number3
  return minimum

print(minimum(12, 27, 36))  # 정수의 최소값
print(minimum(12.3, 45.6, 9.7))  # 실수의 최소값
print(minimum('yellow', 'red', 'orange'))  # string 의 최소값(ascii)
```

```python
# parameter 의 default 값을 정할 수 있음
# 입력해준 parameter 는 첫번째 것부터 대체함
# 따라서 필수 parameter 는 default 값을 지정안하고 앞에다 놓아야 함

def minimum(number1=2, number2=3, number3=4):
  """Return the maximum of three values.""" # function 설명
  print('input numbers:',number1, number2, number3)
  minimum = number1
  if number2 < minimum:
    minimum = number2
  if number3 < minimum:
    minimum = number3
  return minimum

print(minimum())  # 세 parameter 모두 default 값을 사용함
print(minimum(12))  # 첫번째 parameter 만 입력함
print(minimum(12, 27))  # 1,2 번째 parameter 만 입력함
print(minimum(12, 27, 36))  # 세 parameter 모두 입력함
```

```python
# parameter 를 arbitrary argument list 로 줄 수 있음
# C 의 point 활용과 같은 것
def average(*args):
  """Return the average of arbitrary input numbers."""
  print('input numbers:', args)
  return sum(args) / len(args)

print(average(5, 10))            # 2개의 정수를 parameter 로
print(average(12.3, 45.6, 9.7))  # 3개의 실수를 parameter 로
# list 를 parameter 로 넘겨 줄때는 '*' 를 붙인다. C 의 pointer 와 같음
grades = [98, 76, 71, 87, 83, 90, 57, 79, 82, 94]
print(average(*grades))
```
