```python
"""File, Port 를 통한 Data 의 일렬 흐름"""

# File 이란 Data 가 file descriptor 라고 불리는 port 를 통해 일렬로 흐르는 것을 말한다. 끝은 EOF (End of File) 로 표시된다.
# File 전송에 필요한 것은 serial data 흐름의 방향 (read/write), port(file descriptor), 그리고 target file 이름의 세가지 정보이다.

# 이렇게 read, write 되는 file 은 크게 ascii/unicode 형태의 text file 과, 이진수 형태의 binary file (image 등)로 나누어 진다. 그리고 이 file 들의 해독 방법은  extension 에 따라서 달라진다.

# program 이 시작되면 기본적으로 세개의 port(file descriptor) 가 만들어진다 (standard file objects)
#  sys.stdin - input 명령어로 받는 입력 (keyboard)
#  sys.stdout - print 명령어 출력
#  sys.stderr - error 발생시 message 를 보낸다. 
```

```python
"""File 로 Data 를 보내기 (write)"""

import sys
# data 흐름 = write, port(file descriptor) =  write_test, target file name = account.txt
# with 을 쓰면 data 전송을 끝낸 후 자동으로 'accounts.txt' file 을 닫아주고, 'write_test' 도 닫아주어 편리하다. (C 는 이 과정 별도로 해주어야 함)
with open('accounts.txt', mode='w') as write_test:
    write_test.write('100 Jones 24.98\n')
    write_test.write('200 Doe 345.67\n')
    write_test.write('300 White 0.00\n')
    write_test.write('400 Stone -42.16\n')
    print('500 Rich 224.62', file=write_test) # print 문을 이렇게 쓰면 출력을 monitor 가 아닌, write_test 로 내보낸다.
    print('010 1234 5678', file=sys.stdout) # print 문을 이렇게 쓰면 출력을 일반 print 문 처럼 monitor 로 보낸다.
```

```python
"""File 로 Data 를 받기 (read)"""

# data 흐름 = read, port(file descriptor) =  read_test, target file name = account.txt
with open('accounts.txt', mode='r') as read_test:
    print(f'{"Account":<10}{"Name":<10}{"Balance":>10}')
    for record in read_test:
        account, name, balance = record.split()
        print(f'{account:<10}{name:<10}{balance:>10}')
```
