```python
"""Exception Handling"""
# 'ZeroDivisionError', 'ValueError' 등의 error 가 발생하면 program flow 가  멈춘다. 이런 상황을 exception 이라고 한다.
# Exceltion 이 발생할 수 있는 곳은 알 수 있는 경우가 많다. 따라서 그곳에 exception 이 발생할 경우 어떻게 처리를 하고, program flow 를 계속 할 것인지를 지정할 수 있다.

while True:
    try: # exception 을 일으킬 수 있는 code 들을 모아놓음
        number1 = int(input('Enter numerator: '))
        number2 = int(input('Enter denominator: '))
        result = number1 / number2
    except ValueError: # 입력값에 문제가 있을때 처리
        print('You must enter two integers\n')
    except ZeroDivisionError: # 0으로 나누는 경우가 생길때의 처리
        print('Attempted to divide by zero\n')
    else: # 문제가 없을때의 처리
        print('No exceptions are occured\n')
        print(f'{number1:.3f} / {number2:.3f} = {result:.3f}')
    finally: # 마무리 code 들을 모아 놓음 (file 닫기 등)
        print('Loop is finished\n')
        break
```

```python
# 없는 file 을 open 할때의 error 를 처리하는 방법

try:
    with open('nofile_accounts.txt', mode='r') as read_test:
        print(f'{"Account":<10}{"Name":<10}{"Balance":>10}')
        for record in read_test:
            account, name, balance = record.split()
            print(f'{account:<10}{name:<10}{balance:>10}')
except FileNotFoundError:
    print('The file name you specified does not exist')
    
```
