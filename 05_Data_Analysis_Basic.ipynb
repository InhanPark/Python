```python
"""
Data (Collection of Values) 가 주어지면 일단 기본적으로 그 data 의 특성을 파악하여 서술해 보아야 한다.
이런것을 하는 통계학을 'Descriptive Statistics' 라고 한다.

Data Science (Business Analytics 에서도) 의 가장 기본이다.
"""

# data 특성 파악의 가장 기본은 minimum, maximum, range, count, sum 등이다.

# if 문을 이용하여 세 값의 minimum 을 구해보자
number1 = 5
number2 = 9
number3 = 3

minimum = number1

if number2 < minimum:
  minimum = number2

if number3 < minimum:
  minimum = number3

print('Minimum value is', minimum)
```

```python
# python 에서는 기본 함수 (built-in function) 으로 min, max 를 제공한다.
print('Minimum value is', min(number1, number2, number3))
print('Maximum value is', max(number1, number2, number3))
print('Value Range is', min(number1, number2, number3), '-', max(number1, number2, number3))
```
