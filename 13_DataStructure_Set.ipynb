```python
"""수학의 집합과 같은 Data Structure, Set"""
# Set 은 순서가 없는, 같은 것이 없는, 변경할 수 없는 object 들의 모임이다.

# Ser 은 curly brace 로 만든다
colors = {'red', 'orange', 'yellow', 'green', 'red', 'blue'}
print(type(colors), colors)

# 아래와 같이 출력된다. 중복되었던 'red' 가 없어졌다.
# <class 'set'> {'orange', 'yellow', 'blue', 'red', 'green'}

# 길이를 확인하려면
print(len(colors))

# 어떤 item 이 있는지 확인하려면
print('red' in colors)
print('purple' in colors)
print('purple' not in colors)

# Set 으로 iteration 하기
for color in colors:
    print(color.upper(), end=' ')
    
# built-in function 'set' 으로 만들기
numbers = list(range(10)) + list(range(5))
print(numbers)
print(set(numbers)) # 중복 item 이 자동으로 없어졌다.

# empty set 만들기
a = set()  # '{} 는 dictionary 만들때 사용
print(type(a), a)
```

```python
"""수학의 집합과 같은 연산을 한다"""

# 집합의 비교, 원소들이 같으면 True, 다르면 False
print({1, 3, 5} == {3, 5, 1})
print({1, 3, 5} == {3, 5, 2})
print({1, 3, 5} != {3, 5, 1})

# 부분집합(subset)이면 True, 아니면 Flase
print({1, 3, 5} < {3, 5, 1})
print({1, 3, 5} < {7, 3, 5, 1})
print({1, 3, 5} <= {3, 5, 1})
print({1, 3} <= {3, 5, 1})

# subset 확인 method
print({1, 3, 5}.issubset({3, 5, 1}))
print({1, 3}.issubset({3, 5, 1}))
print({1, 3}.issubset({7, 5, 1}))

# superset 확인 method
print({1, 3, 5}.issuperset({3, 5, 1}))
print({1, 3, 5}.issuperset({3, 1}))
print({1, 7, 5}.issuperset({3, 1}))
```

```python
# 합집합(union) 만들기
union_a = {1, 3, 5} | {2, 3, 4}
print(union_a)
union_b = {1, 3, 5}.union([20, 30, 40])
print(union_b)

# 교집합(intersection) 만들기
isect_a = {1, 3, 5} & {2, 3, 4}
print(isect_a)
isect_b = {1, 3, 5}.intersection([1, 2, 2, 3, 3, 4, 4])
print(isect_b)

# 차집합(difference) 만들기
diff_a = {1, 3, 5} - {2, 3, 4}
print(diff_a)
diff_b = {1, 3, 5, 71}.difference([2, 2, 3, 3, 4, 4])
print(diff_b)

# 여집합/대칭차집합(symmetric difference) 만들기
symdiff_a = {1, 3, 5} ^ {2, 3, 4}
print(symdiff_a)
symdiff_b = {1, 3, 5, 71}.symmetric_difference([2, 2, 3, 3, 4, 4])
print(symdiff_b)

# 서로소(disjoint) 판단하기
print({1, 3, 5}.isdisjoint({2, 4, 6}))
print({1, 3, 5}.isdisjoint({1, 4, 6}))
```

```python
"""집합에 원소를 추가하고 빼기"""
# union operator ('|') 를 '=|' 사용하면 왼쪽 operand 를 변경시킨다.
set_number = {1, 3, 5}
set_number |= {2, 3, 4}
print(set_number)

# update method 를 사용하여도 같은 결과 얻음
set_number.update(range(10))
print(set_number)

# intersection, difference, symmetric difference 도 같은 기능 있음
# '&=', '-=', '^='
# 'intersection_update', 'difference_update', 'symmetric_difference_update'

# 원소 추가하기
set_number.add(17)
set_number.add(3)
print(set_number)

# 원소 빼내기
set_number.remove(3)
print(set_number)

# 제일 앞의 원소 빼내기
set_number.pop()
print(set_number)

# 모두 지우기
set_number.clear()
print(set_number)
```
