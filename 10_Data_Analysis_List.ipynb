```python
import statistics  # statistics module 은 통계분석을 위한 다양한 함수를 제공한다.

# data 를 list 에 넣어서 분석하면 매우 편하다
scores = [85, 93, 45, 89, 85]
s = sum(scores)  # list 에 있는 숫자들의 총합을 구해 준다
l = len(scores)  # list 에 있는 숫자들의 갯수를 구해 준다
print('sum: ', s, 'count: ', l)

# 위 data 모음을 서술할때 가장 기본은 평균치이다
print('average: ', s/l)

# statistics 의 function 들을 사용하면 여러 종류의 평균을 쉽게 구할 수 있다
print('mean: ', statistics.mean(scores))   # 평균값
print('median: ', statistics.mean(scores)) # 중앙값
print('mode: ', statistics.mean(scores))   # 최빈값

# list 안에 있는 숫자들을 한번에 sorting 도 할 수 있다
print(sorted(scores))
```
