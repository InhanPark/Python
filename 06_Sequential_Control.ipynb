```python
# 예전에 goto 문은 code 를 엉망으로 만드는 원흉으로 지목됨.
# Code 의 흐름을 아무곳으로 튀게 만들기 때문에...
# 그래서 goto 없이도 code 를 만들 수 있다는 연구결과가 나왔음
# sequential execution (if 문)과 repetition statement (for, while)만으로
# code 를 작성하면 자연히 structured programming 이 된다는 의미였음
# "structured programming = goto elimination"
# python 은 goto 자체가 없음
```

```python
# 한가지 조건만 check 할때
grade = 65

if grade >= 60:
  print('Passed')
```

```python
# 두가지 조건을 check 할때
grade = 55

if grade >= 60:
  print('Passed')
else:   # 조건에 맞지 않는 경우는 모두 이렇게 처리
  print('Failed')
```

```python
# 여러 조건을 check 할때 (위에서부터 차례로)
grade = 75

if grade >= 90:
  print('A')
elif grade >= 80:
  print('B')
elif grade >= 70:
  print('C')
else:   # 조건에 맞지 않는 경우는 모두 이렇게 처리
  print('Failed')
```
