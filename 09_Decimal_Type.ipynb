```python
"""돈계산용 특수 Data Type, Decimal"""

# 소수점 아래에서도 정확히 떨어지는 숫자가 필요한 경우 float data type 으로는 안된다

from decimal import Decimal  # library 사용이 필요함

# float type 의 문제점. 소수점이하 20자리를 표시해보면 문제가 있다. 거액을 다룰때는 문제가 된다.
amount = 112.31
print(amount)
print(f'{amount:.20f}')

# decimal 을 사용하면 마치 정수를 다룰때처럼 깔끔하다
d_amount = Decimal('112.31')
print(f'{d_amount:.20f}')
```

```python
# Decimal 은 특수 data type 이고, 사칙연산이 가능하다

x = Decimal('10.5')
y = Decimal('2')

print( x + y)
print( x // y)

type(x)
```
